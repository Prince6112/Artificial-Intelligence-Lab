## A* search Algorithm
A* algorithm works based on heuristic methods and this helps achieve optimality. A* is a different form of the best-first algorithm
If The f(n) denotes the cost, A* chooses the node with the lowest f(n) value. Here ‘n’ denotes the neighbouring nodes. The calculation of the value can be done as shown below:

f(n)=g(n)+h(n)

g(n) = shows the shortest path’s value from the starting node to node n

h(n) = The heuristic approximation of the value of the node

```python
#Initialising the graph class
class graph:
    
    graph_dict={}
    
    #function to add edges to the graph
    def addedge(self,node,neighbour,gn,fn):
        if node not in self.graph_dict:
            self.graph_dict[node]=[[neighbour,gn,fn]]
        else:
            self.graph_dict[node].append([neighbour,gn,fn])

    #Function to compute the A * Search algorith,
    def AStarSearch(self,source,target):
        visited={}
        bfspath=[]
        
        for i in self.graph_dict:
            visited[i]=False
            for j in self.graph_dict[i]:
                visited[j[0]]=False
        
        #putting the nodes to explore in bfspath and sorting it based on f(n)
        bfspath=self.graph_dict[source]
        
        #sorting the list based on f(n)=h(n)+g(n)
        bfspath.sort(key = lambda x: x[1]+x[2])
        
        #looping till the answer is found
        while(bfspath!=[]):
            path=bfspath.pop(0)
            visited[path[0]]=True
            
            print("->",path[0],end="")
            
            if path[0]==target:
                print("\nTarget found")
                break
            try:
                for node in self.graph_dict[path[0]]:
                    if visited[node[0]]!=True:
                        bfspath.append(node)
                bfspath.sort(key = lambda x: x[1]+x[2])
            except:
                pass         
    

    def display(self):
        for node in self.graph_dict:
            print(node,self.graph_dict[node])
    
```

```python
g=graph()
g.addedge('S','A',7,9)
g.addedge('S','B',2,7)
g.addedge('S','C',3,8)
g.addedge('A','D',4,8)
g.addedge('A','B',3,7)
g.addedge('B',"D",4,7)
g.addedge('D','F',5,6)
g.addedge('B','H',1,7)
g.addedge('H','F',3,6)
g.addedge('H','G',2,3)
g.addedge('C','L',2,4)
g.addedge('L','I',4,4)
g.addedge('L','J',4,4)
g.addedge('K','E',5,0)
g.addedge('G','E',2,0)
g.addedge('I','K',4,3)
g.display()
```

```python
g.AStarSearch('S','E')
```

## Space Complexity - BFS
The total space occupied by the program is the space required to store each edges, vertices and the priority between them. That is O(n), where N is the number of nodes in the graph.

## Time Complexity - BFS
The worst case time complexity for BFS is O(N * log N) where N here is the number of nodes in the graph. The sorting of the list is done in O(log N) time and in worst case the program needs to travel to all the nodes hence O(N). Combining them both we get O(N * log N) time complexity
