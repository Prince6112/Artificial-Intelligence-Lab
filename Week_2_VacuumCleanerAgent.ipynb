```python
room=input("Enter the room which Vacuum Cleaner is currently is in (A/B): ")
if room=='A':
    otherroom='B'
else:
    otherroom='A'
trashroom=input("Enter the number of trash piles in the "+room+" room: ")
trashotherroom=input("Enter the number of trash piles in other room: ")
```

```python
#Cleaning the room which the vacuum is initially in
print("The Vacuum cleaner is currently in '"+room+"' Room")
print("Found "+trashroom+" piles of garbage in the '"+room+"' Room")

#Cleaning the trash of the room
print("\nCleaning them...")
for x in range(int(trashroom),0,-1):
    print("Trash",x,"-Cleaned")

print("Cleaned Room "+room)

#Cleaning the next room
print("\nMoving to the next room")

print("\nThe Vacuum cleaner is currently in '"+otherroom+"' Room")
print("Found "+trashotherroom+" piles of garbage in the '"+room+"' Room")

#Cleaning the other room
print("\nCleaning them...")
for x in range(int(trashotherroom),0,-1):
    print("Trash",x,"-Cleaned")
print("Cleaned Room "+otherroom)


#Both the rooms are now clean    
print("\nBoth the rooms A and B are now clean")

#Calculating the total path cost, which is sum of the trash cleaned plus one for movement
print("\nThe Total path cost is the number of actions by the vacuum which is ",int(trashroom)+int(trashotherroom)+1)
```
